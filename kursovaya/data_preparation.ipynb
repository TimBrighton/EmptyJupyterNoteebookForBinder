```python
# Загрузка зависимостей
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Загрузка и анализ набора данных
country_dataset = pd.read_csv('Набор_3_страны_мира.csv',  sep=';')
country_dataset.head(10)
```

```python
# Создаем набор данных, в котором будут храниться обработанные данные
dataset = pd.DataFrame()
```

```python
# обработка колонки "region"
data = country_dataset['region']
data = pd.get_dummies(data)
data = np.array([data[i[1]] * (i[0]+1) for i in  enumerate(data)]).flatten()
data = data[data != 0]
plt.hist(data, bins = 50)
plt.show()
```

```python
plt.hist(data**0.5, bins = 50)
plt.show()
plt.hist(np.log(data), bins = 50)
plt.show()
```

```python
data = np.array(data**0.5).reshape(-1, 1)
dataset['region'] =MinMaxScaler().fit_transform(data).flatten()
dataset.head(10)
```

```python
# обработка колонки "population"
data = country_dataset['population']
plt.hist(data, bins = 50)
plt.show()
```

```python
data = data.replace(0, data.mean())
plt.hist(data, bins = 50)
plt.show()
```

```python
plt.hist(data**0.5, bins = 50)
plt.show()
plt.hist(np.log(data), bins = 50)
plt.show()
```

```python
data = np.array(np.log(data)).reshape(-1, 1)
dataset['population'] =MinMaxScaler().fit_transform(data).flatten()
dataset.head(10)
```
